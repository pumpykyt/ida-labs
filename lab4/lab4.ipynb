```python
country_data4 = country_data4[country_data4['gdpp'] < 55000]
country_data4 = country_data4[country_data4['child_mort'] < 140]
country_data4 = country_data4[country_data4['income'] < 50000]
country_data4.info()
```

```python
data1 = list(zip(country_data4["gdpp"], country_data4["income"]))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data1)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data1)
fig, ax = plt.subplots(figsize=(8, 6))
plt.scatter(country_data4["gdpp"], country_data4["income"], c=kmeans.labels_, cmap='Spectral')
ax.set_xlabel('GDPP', fontsize=20)
ax.set_ylabel('INCOME', fontsize=20)
plt.show()
```

```python
data2 = list(zip(country_data4["gdpp"], country_data4["child_mort"]))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data2)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kl = KneeLocator(range(1, 11), inertias, curve="convex", direction="decreasing")
kl.plot_knee()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data2)
customcmap = ListedColormap(["crimson", "mediumblue", "darkmagenta"])
fig, ax = plt.subplots(figsize=(8, 6))
plt.scatter(country_data4["gdpp"], country_data4["child_mort"], c=kmeans.labels_, cmap = customcmap)
ax.set_xlabel('GDPP', fontsize=14)
ax.set_ylabel('Child Deaths', fontsize=14)
plt.show()
```

```python
X = country_data4[['child_mort', 'income']].copy()
from sklearn.mixture import GaussianMixture
n_clusters = 3
gmm_model = GaussianMixture(n_components=n_clusters)
gmm_model.fit(X)
cluster_labels = gmm_model.predict(X)
X = pd.DataFrame(X)
X['cluster'] = cluster_labels
```

```python
fig, ax = plt.subplots(figsize=(9, 7))
sns.scatterplot(x='child_mort', y='income', data=X, hue='cluster', palette="deep", s=70, ax = ax) 
plt.title('Clusters Identified by Guassian Mixture Model')
plt.xlabel('Child Deaths')
plt.ylabel('Income')
```

```python
country_data5 = country_data4[["country"]]
country_data5=pd.merge(country_data5,X, left_index=True,right_index=True)
country_data5.nsmallest(8, ['income', 'child_mort'])
```

```python
country_data5['cluster'].value_counts(normalize=True)
```

```python
cluster_child_mort=pd.DataFrame(X.groupby(["cluster"]).child_mort.mean())
cluster_income=pd.DataFrame(X.groupby(["cluster"]).income.mean())
country_data6 = pd.concat([Cluster_child_mort,Cluster_income], axis=1)
country_data6.columns = ["child_mort","income"]
country_data6
```

```python
fig = plt.figure(figsize = (10,6))
country_data6.rename(index={0: 'Under-developed Countries'},inplace=True)
country_data6.rename(index={1: 'Developing Countries'},inplace=True)
country_data6.rename(index={2: 'Developed Countries'},inplace=True)
s=sns.barplot(x=country_data6.index,y='income',data=country_data6)
plt.xlabel('Country Groups', fontsize=10)
plt.ylabel('Income per person', fontsize=10)
plt.title('Country Groups On the basis of INCOME')
plt.show()
```
